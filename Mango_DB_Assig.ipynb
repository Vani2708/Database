# **Mango DB Theoretical questions:**

1. **What are the key differences between SQL and NoSQL databases?**

   SQL uses structured tables with fixed schema; NoSQL uses flexible schema and stores data in formats like documents or key-value.

2. **What makes MongoDB a good choice for modern applications?**

   It offers flexibility, scalability, high performance, and handles large unstructured data easily.

3. **Explain the concept of collections in MongoDB.**

   A collection is a group of documents, similar to a table in SQL.

4. **How does MongoDB ensure high availability using replication?**

   It uses replica sets with primary and secondary nodes to duplicate data and ensure failover.

5. **What are the main benefits of MongoDB Atlas?**

   Cloud hosting, automatic scaling, backups, security, and easy management.

6. **What is the role of indexes in MongoDB, and how do they improve performance?**

   Indexes speed up queries by reducing the amount of data scanned.

7. **Describe the stages of the MongoDB aggregation pipeline.**

   Common stages include `$match`, `$group`, `$project`, `$sort`, and `$limit`.

8. **What is sharding in MongoDB? How does it differ from replication?**

   Sharding splits data across servers; replication copies data for backup and availability.

9. **What is PyMongo, and why is it used?**

   PyMongo is a Python library used to connect and work with MongoDB.

10. **What are the ACID properties in the context of MongoDB transactions?**

    Atomicity, Consistency, Isolation, Durability ensure reliable and safe transactions.

11. **What is the purpose of MongoDB’s explain() function?**

    It shows query execution details and helps optimize performance.

12. **How does MongoDB handle schema validation?**

    It uses JSON schema to enforce rules on document structure.

13. **What is the difference between a primary and a secondary node in a replica set?**

    Primary handles writes; secondary replicates data and supports reads.

14. **What security mechanisms does MongoDB provide for data protection?**

    Authentication, authorization, encryption, and auditing.

15. **Explain the concept of embedded documents and when they should be used.**

    Embedded documents store related data together; used when data is frequently accessed together.

16. **What is the purpose of the $lookup stage in aggregation?**

    It joins data from another collection, similar to SQL joins.

17. **What are some common use cases for MongoDB?**

    Real-time apps, big data, IoT, content management, and e-commerce.

18. **What are the advantages of using MongoDB for horizontal scaling?**

    It distributes data across multiple servers, improving performance and scalability.

19. **How do MongoDB transactions differ from SQL transactions?**

    MongoDB supports multi-document transactions but is mainly optimized for single-document operations.

20. **What are the main differences between capped collections and regular collections?**

    Capped collections have fixed size and overwrite old data; regular collections grow dynamically.

21. **What is the purpose of the $match stage in MongoDB’s aggregation pipeline?**

    It filters documents, similar to WHERE in SQL.

22. **How can you secure access to a MongoDB database?**

    Use authentication, role-based access, IP whitelisting, and SSL/TLS encryption.

23. **What is MongoDB’s WiredTiger storage engine, and why is it important?**

    It is the default engine that provides better performance, compression, and concurrency control.


# **Practical Questions**

```python
import pandas as pd
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["superstore"]
collection = db["orders"]

df = pd.read_csv("Superstore.csv")
data = df.to_dict("records")

collection.insert_many(data)
```

```python
for doc in collection.find():
    print(doc)
```

```python
print(collection.count_documents({}))
```

```python
for doc in collection.find({"Region": "West"}):
    print(doc)
```

```python
for doc in collection.find({"Sales": {"$gt": 500}}):
    print(doc)
```

```python
for doc in collection.find().sort("Profit", -1).limit(3):
    print(doc)
```
